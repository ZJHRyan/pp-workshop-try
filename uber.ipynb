```python
import pandas as pd
```

```python
path='data/restaurants.csv'
df_data=pd.read_csv(path,encoding ='UTF-8')
df=df_data.copy()
df
```

```python
initial_data=df.dropna(axis=0,how='any')
initial_data=initial_data.replace(to_replace ="$", value =1)
initial_data=initial_data.replace(to_replace ="$$", value =2)
initial_data=initial_data.replace(to_replace ="$$$", value =3)
initial_data=initial_data.replace(to_replace ="$$$$", value =4)
initial_data
```

```python
initial_data.describe()
```

```python
import matplotlib.pyplot as plt
initial_data.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
plt.rcParams['figure.figsize']=(12.8, 7.2)
initial_data.plot(kind="scatter", x="lng", y="lat",alpha=0.01)
```

```python
plt.rcParams['figure.figsize']=(12.8, 7.2)
initial_data.plot(kind="scatter", x="lng", y="lat", alpha=0.4,
    s=initial_data["score"]/500, label="score",
    c="ratings", cmap=plt.get_cmap("jet"), colorbar=True,
)
plt.legend()
```

```python
corr_matrix = initial_data.corr()
corr_matrix
```

```python
text=initial_data.category.str.split(', ',expand=True).stack().value_counts()
text
```

```python
import wordcloud
import matplotlib.pyplot as plt
# Create and generate a word cloud image:
wordcloud = wordcloud.WordCloud(max_font_size=50, max_words=100, background_color="white").generate_from_frequencies(text)
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
